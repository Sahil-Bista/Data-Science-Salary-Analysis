●	Write a Python program to show summary statistics of sum, mean, standard deviation, skewness, and kurtosis of any chosen variable.

*Let the chosen variable be salary_in_usd*


```python
#iterates over each rows of the salary_in_usd column and adds all the values present and stores it in a variable sum
sum = 0
for each in salary_data.salary_in_usd:
  sum += each
print(sum)
```

```python
#to check if the sum calculates using hardcore python was right
salary_data.salary_in_usd.sum()
```

MEAN

```python
# to find the number of values present in the salary_in_usd column
length = len(salary_data.salary_in_usd)
print(length)
```

```python
# here sum = sum of all the salaries_in_usd as previously found
total_salaries_in_usd = sum
print(total_salaries_in_usd)
```

```python
mean = total_salaries_in_usd / length
print(mean)
```

```python
#to check if the mean calculated using hardcore python was right
salary_data.salary_in_usd.mean()
```

Standard Deviation


```python
#calculating the numerator for standard deviation which is the (sum of individual data in each row - mean)^2
numerator = 0
for each in salary_data.salary_in_usd:
  sub = (each - mean)**2
  numerator += sub
print(numerator)
```

```python
standard_deviation = (numerator/length)**(1/2)
print(standard_deviation)
```

```python
#checking if the standard deviation calculated was corrcetv using inbuilt function
salary_data.salary_in_usd.std()
```

Skewness


```python
length = len(salary_data.salary_in_usd)
print(length)
```

```python
#so the number of terms in our case is in odd number , using the formula for median:
numerator_part = 0
for numbers in salary_data.salary_in_usd:
  numerator_part += (numbers-mean)**3
print(numerator_part)
```

```python
print(numerator_part)
```

```python
#calculating the denominator for the formula used to calculate skewness
denominator = (length - 1 )*(standard_deviation)**3
print(denominator)
```

```python
#calculating skewness
skewness = numerator_part/denominator
print(skewness)
```

```python
#using in built function to check if the calculated skewness was correct
salary_data.salary_in_usd.skew()
```

Kurtosis


```python
kurtosis_num = 0
for kurt in salary_data.salary_in_usd:
  kurtosis_num += ((kurt-mean)/standard_deviation)**4

print(kurtosis_num)
```

```python
#calculating the first expression in kurtosis formula
first_exp = (length*(length+1))/((length - 1) * (length -2) * (length -3))*(kurtosis_num)
print(first_exp)
```

```python
#calculating the second expression in kurtosis formula
second_exp =  ((3*(length - 1)**2))/((length-2)*(length-3))
print(second_exp)
```

```python
#calculaitng kurtosis
kurtosis = first_exp - second_exp
print(kurtosis)
```

```python
#checking if the value for kurtosis was right
salary_data.salary_in_usd.kurt()
```

Write a python program to find out top 15 jobs. Make a bar graph of sales as well.

```python
#calculating the job titles that have been assigned the most or basically  calculating top 15 jobs based upon the repition of values fot that specific job_title
job_counts = salary_data['job_title'].value_counts().head(15)
```

```python
#showing the top 15 job titles in a bar
mpl.figure(figsize=(12,7))
job_counts.plot(kind='bar',color='lightblue')
mpl.xlabel("Job Title")
mpl.ylabel('Frequency')
mpl.title("Top 15 job titles")
mpl.show()
```

2) What job has the highest salaries? Illustrate with bar graph


```python
#groups the dataframe in the basis of job titles , calculates the avrage salary for that job title and then sorts them in descendinng order that is from highest to lowest
#the head(7) filters the top 7b rows of the reulted dataframe , meaning top 7 hihgest salary jobs
highest_pay_job = salary_data.groupby('job_title').salary_in_usd.mean().sort_values(ascending=False).head(7)
mpl.bar(highest_pay_job.index,highest_pay_job.values, width=0.3)
mpl.xlabel("Job Title")
mpl.ylabel("salary_in_usd")
mpl.xticks(rotation = 90 )
mpl.title('Highest salary jobs')
mpl.show()

```

Wriye a python program to find out salaries based on experience level. Illustrate it through bar graph.

```python
#basically groups the jobs based on experience level, calculates the mean salary for that group and makes a bar graph using the experience level and salary_in_usd
salaries_on_experience = salary_data.groupby('experience_level')['salary_in_usd'].mean()
mpl.bar(salaries_on_experience.index,salaries_on_experience.values,width=0.4)
mpl.xticks(rotation = 90)
mpl.xlabel("Experience Level")
mpl.ylabel("salary_in_usd")
```
